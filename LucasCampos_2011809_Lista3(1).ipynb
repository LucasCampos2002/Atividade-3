```python
#1) Criando o dataframe
import pandas as pd
from google.colab import drive
drive.mount('/drive')

df = pd.read_csv('/drive/MyDrive/class_bank_marketing.csv')
display(df)
```

```python
#2) Apagando todos os nulos da coluna previous
df.drop(df[df['previous'] == 0].index, inplace = True)
display(df['previous'])
```

```python
#3) Calculando a média da coluna previous
media = df['previous'].mean()
print(media)
```

```python
#4) Eliminar linhas que possuam valores muito distantes da média
# muito distantes da media = mais que o dobro da media

out = df[df['previous'] > media * 2].index

new = df

new.drop(out, inplace = True)

display(new)
```

```python
#5) Criando um novo df que contenha apenas 5 colunas do conjunto anterior

newdf = new[:]

final = newdf[['age','job','month','duration','previous']]

display(final)
```

```python
#6)  (Coluna A - Coluna B)*número aleatório (entre 0 e 1).
# neste caso coluna A = age & coluna B = previous
import random
aleatorio = random.randint(0,1000)
numero = aleatorio/1000
# print(numero)

df['colunaNova'] = ((df['age'] - df['previous']) * numero)

display(df)
```

```python
#7) Embaralhando aleatoriamente 
import numpy as np 

novo = df.reindex(np.random.permutation(df.index))

display(novo)
```

```python
#8) df ordenado by balance (crescente)

df = df.sort_values(by='balance', ascending = True)
display(df)
```
